```python
def actualizar_w_y_b(x,y,w,b,alpha):
    dl_dw = 0.0
    dl_db = 0.0
    N=len(y)
    
    for i in range(N):
        
        dl_dw += -2*x[i]*(y[i]-(w*x[i]+b))
        
        dl_db += -2*(y[i]-(w*x[i]+b))
        
    #Actualizar w y b
    
    w=w-(1/float(N))*dl_dw*alpha
    
    #print(w)
    
    b=b-(1/float(N))*dl_db*alpha
    
    return w,b
```

```python
# La función que recorre varias épocas se muestra a continuación:
```

```python
def train(y,x,w,b,alpha,epochs):
    for e in range(epochs):
        w,b=actualizar_w_y_b(y,x,w,b,alpha)
        
        #Registrar el progreso
        if e%400.00 ==0:
            print("epoch",e,"pérdida promedio:",avg_loss(y,x,w,b),"w",w,"b",b)
        
       
    return w,b
```

```python
def plot_DG(y,x,w,b,alpha,epochs):
    
    for e in range(epochs):
        
        w,b=actualizar_w_y_b(y,x,w,b,alpha)
        
        if e%400.00 ==0:
            x = x
            y1 = y
            y2 = w*x+b
            
            plt.scatter(x,y1)
            plt.plot(x,y2,c="red")
            
            plt.legend(["Modelo", "Datos"])
            
            plt.title("Epoch:%d" %e)
            p=plt.show()
        
    return p    
```

```python
def avg_loss(y,x,w,b):
    N=len(y)
    error_total=0.00
    for i in range(N):
        error_total += (x[i]-(w*y[i]+b))**2
    return error_total/float(N)    
```

```python
plot_DG(df["National-US"],df["MA-Boston"],0.00,0.00,0.0000001,3000)
```

```python
train(df["National-US"],df["MA-Boston"],0.00,0.00,0.0000001,3000)
```
